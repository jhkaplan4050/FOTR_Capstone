```python
# Import for dataframe package
import pandas as pd
```

```python
# Read in the raw data and view 
ft = pd.read_csv('FOTR_Capstone-main/original_data/FROGS_AND_TOADS.csv')
ft.head()
```

```python
frogs = ['WoodFrog', 'ChorusFrog', 'SpringPeeper', 'AmericanToad', 'LeopardFrog', 
         'GrayTreeFrog', 'GreenFrog', 'BullFrog']
names = ['Wood Frog', 'Chorus Frog', 'Spring Peeper', 'American Toad', 'Leopard Frog', 
         'Gray Tree Frog', 'Green Frog', 'Bull Frog']

# Make dataframe separated by frog type instead of observation
df = pd.DataFrame()
trsq = []
yr = []
full_frogs = []
# For each row, find all rows where that frog was 'True' (1) and add that row into the new dataframe
# Collect all the data into arrays 
for i in range(len(frogs)):
    trsq.append(ft[ft[frogs[i]] == 1]['TRSQ'])
    yr.append(ft[ft[frogs[i]] == 1]['Observationyear'])
    for j in range(len(ft[ft[frogs[i]] == 1]['TRSQ'])):
        full_frogs.append(names[i])
df['frog_type'] = full_frogs
trsq_full = []
yr_full = []
for i in range(len(trsq)):
    for j in range(len(trsq[i])):
        trsq_full.append(trsq[i].iloc[j])
        yr_full.append(yr[i].iloc[j])
# Add arrays into the new dataframe
df['TRSQ'] = trsq_full
df['Observationyear'] = yr_full
df['Value'] = 1
# Drop any repeats so its one entry per frog, trsq and year 
df = df.drop_duplicates()
df
```

```python
# Save as CSV file to input 
df.to_csv('final_ft.csv')
```
