```python
ft_types = ft[['WoodFrog', 'ChorusFrog', 'SpringPeeper', 'AmericanToad', 'LeopardFrog', 
               'GrayTreeFrog', 'GreenFrog', 'BullFrog']]
ft['num_species'] = ft_types[ft_types == True].sum(axis=1).astype(int)
```

```python
ft[['WoodFrog', 'ChorusFrog', 'SpringPeeper', 'AmericanToad', 'LeopardFrog', 
               'GrayTreeFrog', 'GreenFrog', 'BullFrog']] = ft[['WoodFrog', 'ChorusFrog', 'SpringPeeper', 'AmericanToad', 'LeopardFrog', 
               'GrayTreeFrog', 'GreenFrog', 'BullFrog']].astype(int)
```

```python
num_year = ft[['Observationyear', 'num_species']].groupby('Observationyear').mean()
num_year
```

```python
yr_entries = ft['Observationyear'].value_counts().sort_index().tolist()
```

```python
ft_types = ft[['Observationyear', 'WoodFrog', 'ChorusFrog', 'SpringPeeper', 'AmericanToad', 'LeopardFrog', 
               'GrayTreeFrog', 'GreenFrog', 'BullFrog']]
by_year = ft_types.groupby('Observationyear').sum()
by_year['avg_species'] = num_year['num_species']
by_year['num_entries'] = yr_entries
by_year['WoodFrog'] = by_year['WoodFrog'] / by_year['num_entries']
by_year['ChorusFrog'] = by_year['ChorusFrog'] / by_year['num_entries']
by_year['SpringPeeper'] = by_year['SpringPeeper'] / by_year['num_entries']
by_year['AmericanToad'] = by_year['AmericanToad'] / by_year['num_entries']
by_year['LeopardFrog'] = by_year['LeopardFrog'] / by_year['num_entries']
by_year['GrayTreeFrog'] = by_year['GrayTreeFrog'] / by_year['num_entries']
by_year['GreenFrog'] = by_year['GreenFrog'] / by_year['num_entries']
by_year['BullFrog'] = by_year['BullFrog'] / by_year['num_entries']
by_year
```

```python
by_year.to_csv('normalized_data_by_year.csv')
```

```python
df = pd.DataFrame()
frogs = ['WoodFrog', 'ChorusFrog', 'SpringPeeper', 'AmericanToad', 'LeopardFrog', 
         'GrayTreeFrog', 'GreenFrog', 'BullFrog']
trsq = []
yr = []
full_frogs = []
for frog in frogs:
    trsq.append(ft[ft[frog] == 1]['TRSQ'])
    yr.append(ft[ft[frog] == 1]['Observationyear'])
    for i in range(len(ft[ft[frog] == 1]['TRSQ'])):
        full_frogs.append(frog)
df['frog_type'] = full_frogs
trsq_full = []
yr_full = []
for i in range(len(trsq)):
    for j in range(len(trsq[i])):
        trsq_full.append(trsq[i].iloc[j])
        yr_full.append(yr[i].iloc[j])
df['TRSQ'] = trsq_full
df['Observationyear'] = yr_full
df
```

```python
df = df.drop_duplicates()
```

```python
df.to_csv('frog_type_loc_yr.csv')
```
