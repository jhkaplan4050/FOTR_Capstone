```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

bug_data = pd.read_csv('benthics_updated.csv')




bug_data = bug_data.dropna()
bug_data.head()
```

```python
conditions = [
    (bug_data['SQI'] <19),
    (bug_data['SQI'] >= 19) & (bug_data['SQI'] <= 33),
    (bug_data['SQI'] >=34) & (bug_data['SQI'] <= 48),
    (bug_data['SQI'] > 34)
    ]

# create a list of the values we want to assign for each condition
values = ['Poor', 'Fair', 'Good', 'Excellent']

# create a new column and use np.select to assign values to it using our lists as arguments
bug_data['SQI_Rating'] = np.select(conditions, values)

# display updated DataFrame
bug_data.head()

extracted_col = bug_data["Date"]
```
