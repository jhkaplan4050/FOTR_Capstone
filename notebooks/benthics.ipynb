```python
import pandas as pd
import numpy as np
```

```python
ben_sqi = pd.read_csv('FOTR_Capstone-main/BENTHIC_MACROINVERTEBRATES_SQI.csv')
```

```python
ben_sqi.drop(columns=['Unnamed: 47'], inplace=True)
```

```python
ben_sqi_new = ben_sqi
ben_sqi_new.head()
```

```python
sqi_cols = ['F22 SQI', 'SP22 SQI', 'F21 SQI',
       'SP21 SQI', 'F20 SQI', 'SP20 SQI', 'F19 SQI', 'SP19 SQI', 'F18 SQI',
       'SP18 SQI', 'F17 SQI', 'SP17 SQI', 'F16 SQI', 'SP16 SQI', 'F15 SQI',
       'SP15 SQI', 'F14 SQI', 'SP14 SQI', 'F13 SQI', 'SP13 SQI', 'F12 SQI',
       'SP12 SQI', 'F11 SQI', 'SP11 SQI', 'F10 SQI', 'SP10 SQI', 'F09 SQI',
       'SP09 SQI', 'F08 SQI', 'SP08 SQI', 'F07 SQI', 'SP07 SQI', 'F06 SQI',
       'SP06 SQI', 'F05 SQI', 'SP05 SQI', 'F04 SQI', 'SP04 SQI', 'F03 SQI',
       'SP03 SQI', 'F02 SQI', 'SP02 SQI', 'F01 SQI', 'SP01 SQI']
```

```python
benthics_sqi = pd.DataFrame(columns=['FIELDID', 'Latitude', 'Longitude', 'Date', 'SQI'], index=['FIELDID'])
for index, row in ben_sqi_new.iterrows():
    for col in ben_sqi_new[sqi_cols]:
        if col[0] == 'F':
            # pull out sampling month
            date = '09/15/' + col[1:3]
            # convert to datetime
            date_time = pd.to_datetime(date)
            # pull out SQI value
            sqi = row[col]
            # compile row dict
            info = {'FIELDID':row['FIELDID'], 'Latitude':row['Latitude'], 'Longitude':row['Longitude'], 'Date':date_time, 'SQI':sqi}
            info_df = pd.DataFrame(info, index=['Date'])
            # create new row for each sampling event
            benthics_sqi = pd.concat([benthics_sqi, info_df], ignore_index = True)
```

```python
stoneflies_sqi = pd.DataFrame(columns=['FIELDID', 'Latitude', 'Longitude', 'Date', 'SQI'], index=['FIELDID'])
for index, row in ben_sqi_new.iterrows():
    for col in ben_sqi_new[sqi_cols]:
        if col[0] == 'S':
            # pull out sampling month
            date = '01/15/' + col[2:4]
            # convert to datetime
            date_time = pd.to_datetime(date)
            # pull out SQI value
            sqi = row[col]
            # compile row dict
            info = {'FIELDID':row['FIELDID'], 'Latitude':row['Latitude'], 'Longitude':row['Longitude'], 'Date':date_time, 'SQI':sqi}
            info_df = pd.DataFrame(info, index=['Date'])
            # create new row for each sampling event
            stoneflies_sqi = pd.concat([stoneflies_sqi, info_df], ignore_index = True)
```

```python
benthics_sqi.dropna(inplace=True)
```

```python
benthics_sqi['SQI'] = benthics_sqi['SQI'].astype(int)
```

```python
benthics_sqi['SQI'].value_counts()
```

```python
bins = [0, 19, 33, 48, np.inf]
names = ['Poor', 'Fair', 'Good', 'Excellent']

benthics_sqi['SQI_Rating'] = pd.cut(benthics_sqi['SQI'], bins, labels=names)
```

```python
benthics_sqi.drop_duplicates(inplace=True)
```
