```python
# Import for dataframe package
import pandas as pd
import numpy as np
```

```python
# Read in the raw data and view 
# IF YOU REORGANIZE THE FILES, CHANGE THE PATH IN THE NEXT LINE
ft = pd.read_csv('../original_data/FROGS_AND_TOADS.csv')
ft.head()
```

```python
frogs = ['WoodFrog', 'ChorusFrog', 'SpringPeeper', 'AmericanToad', 'LeopardFrog', 
         'GrayTreeFrog', 'GreenFrog', 'BullFrog']
names = ['Wood Frog', 'Chorus Frog', 'Spring Peeper', 'American Toad', 'Leopard Frog', 
         'Gray Tree Frog', 'Green Frog', 'Bull Frog']

# Make dataframe separated by frog type instead of observation
df = pd.DataFrame()
trsq = []
yr = []
full_frogs = []

# For each row, find all rows where that frog was 'True' (1) and add that row into the new dataframe
# Collect all the data into arrays 
for i in range(len(frogs)):
    trsq.append(ft[ft[frogs[i]] == 1]['TRSQ'])
    yr.append(ft[ft[frogs[i]] == 1]['Observationyear'])
    for j in range(len(ft[ft[frogs[i]] == 1]['TRSQ'])):
        full_frogs.append(names[i])
df['frog_type'] = full_frogs
trsq_full = []
yr_full = []
for i in range(len(trsq)):
    for j in range(len(trsq[i])):
        trsq_full.append(trsq[i].iloc[j])
        yr_full.append(yr[i].iloc[j])
        

# Add arrays into the new dataframe
df['TRSQ'] = trsq_full
df['Observationyear'] = yr_full
df['Value'] = 1

# Drop any repeats so its one entry per frog, trsq and year 
df = df.drop_duplicates()
df
```

```python
# Save as CSV file to input 
df.to_csv('../modified_data/frog_toad/final_ft.csv')
```

```python
# Find the number of sites each frog type was found at
# Group the dataframe by frog type and observation year, summing the number of sites
frog_year_num_sites = df.groupby(['frog_type', 'Observationyear']).sum()
frog_year_sites = frog_year_num_sites.loc['American Toad']

# Fix the names
for name in names:
    full_name = name + ' Sites'
    frog_year_sites[full_name] = frog_year_num_sites.loc[name]

# Fill any empty years with 0 (no observations) and drop the values column 
frog_year_sites = frog_year_sites.fillna(0)
frog_year_sites = frog_year_sites.drop('Value', axis=1)
frog_year_sites
```

```python
frog_year_sites.to_csv('../modified_data/frog_toad/FT_num_sites.csv')
```

```python
# List the frogs/toads found in each site
frogs_persite = []
trsq = []

for site in df['TRSQ'].values:
    # Get frogs as a list
    all_found = np.unique(df[df['TRSQ']==site]['frog_type'].values)
    # Change to a string: 
    add = all_found[0]
    if len(all_found) > 1:
        for frog in all_found[1:]:
            add += ', ' + frog
    frogs_persite.append(add)
    trsq.append(site)

df_persite = pd.DataFrame()
df_persite['TRSQ'] = trsq
df_persite['frog_names'] = frogs_persite
df_persite.head()
```

```python
df_persite.to_csv('../modified_data/frog_toad/frognames_persite.csv')
```
